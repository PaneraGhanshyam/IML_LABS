```python
'''
Practical No.4: Write a NumPy program to implement following operation

to add, subtract, multiply, divide arguments element-wise

Name:panera ghanshyam
Enroll no:216270307072
'''
#method 1
a=np.random.randint(1,20,(3,3))
print("Orignal array:\n",a)
print("\nAfter addition by 5:\n",np.add(a,5))
print("\nAfter subtraction by 5:\n",np.subtract(a,5))
print("\nAfter multiplication by 5:\n",np.multiply(a,5))
print("\nAfter division by 2:\n",np.divide(a,2))

#method 2
a=np.random.randint(1,20,(3,3))
print("\n\nOrignal array:\n",a)
print("\nAfter addition by 5:\n",a+5)
print("\nAfter subtraction by 5:\n",a-5)
print("\nAfter multiplication by 5:\n",a*5)
print("\nAfter division by 2:\n",a/2)
```

```python
'''
Practical No.4: Write a NumPy program to implement following operation

to round elements of the array to the nearest integer

Name:panera ghanshyam
Enroll no:216270307072
'''

a=np.random.randint(100,200,(3,3),dtype="int64")
a=a/3
print("before round:\n",a)
print("\nafter round:\n",np.round(a))
```
