```python
'''
Write a Pandas program to implement following operation

to create a dataframe from a dictionary and
display it

Name:Panera Ghanshyam.
Enrollment No:216270307072.
'''

d={"name":["ghanshyam","harsh","yash"],"eno":[72,72,73]}

df=pd.DataFrame(d)

print(df)
```

```python
'''
Write a Pandas program to implement following operation

to sort the DataFrame first by 'name' in
ascending order

Name:Panera Ghanshyam.
Enrollment No:216270307072.
'''
d={"name":["ghanshyam","yash","harsh"],"eno":[72,73,74]}

df=pd.DataFrame(d)

print("Before sort:\n",df)

df=df.sort_values(by=["name"],ascending=True)
print()
print("After sort:\n",df)
```

```python
'''
Write a Pandas program to implement following operation

to delete the one specific column from the
DataFrame

Name:Panera Ghanshyam.
Enrollment No:216270307072.
'''

print("Original DataFrame:\n",df)

df=df.drop(columns="eno")

print()
print("After Deleting Eno column:\n",df)
```

```python
'''
Write a Pandas program to implement following operation

to write a DataFrame to CSV file using tab
separator

Name:Panera Ghanshyam.
Enrollment No:216270307072.
'''
df.to_csv('/kaggle/working/dataset.csv', index=False, sep="\t")
```
