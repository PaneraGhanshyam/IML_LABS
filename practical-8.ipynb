```python
'''
Write a Pandas program to create a line plot of
the opening, closing stock prices of given
company between two specific dates.

Name:-Panera Ghanshyam.
Enrollment No:-216270307072.
'''
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../input/nifty50-stock-market-data/HDFCBANK.csv")
start_date = pd.to_datetime('2000-01-03')
end_date = pd.to_datetime('2000-12-29')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df2 = df.loc[new_df]
plt.figure(figsize=(20,20))
df2.plot(x='Date', y=['Open', 'Close']);
plt.suptitle('Opening/Closing stock prices of HDFC BANK,\n 03-01-2000 to 29-12-2000', fontsize=12, color='black')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("$ price", fontsize=12, color='black')
plt.show()
```
